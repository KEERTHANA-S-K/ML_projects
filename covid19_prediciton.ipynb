```python
###Handling features to build model
```

```python
data.info()
```

```python
data.head()
```

```python
##handling date feature
```

```python
data["year"]=data["date"].dt.year
```

```python
data["month"]=data["date"].dt.month
```

```python
data["day"]=data["date"].dt.day
```

```python
data.head()
```

```python
data.drop(["date"],axis=1,inplace=True)
```

```python
### handly state feature
```

```python
pos=data.groupby(['state'])['confirmed'].mean().sort_values().index
dict1={key:index for index,key in enumerate(pos,0)}
data["state"]=data['state'].map(dict1)
```

```python
data.info()
```

```python
### rearange the columns
```

```python
cols = data.columns.tolist()
```

```python
cols
```

```python
index=data.columns.get_loc("confirmed")
```

```python
index
```

```python
new_col= cols[0:index] + cols[index+1:] + cols[index:index+1]
```

```python
new_col
```

```python
data=data[new_col]
```

```python
###splitting the data

x=data.drop(['confirmed'],axis=1)
y=y=data['confirmed']
```

```python
# lets perform train test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# lets perform standardization

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python

# lets create a simple Linear Regression Model
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
model1.fit(x_train, y_train)

y_pred1 = model1.predict(x_test)
```

```python
# lets check the Model accuracy
from sklearn.metrics import r2_score, mean_squared_error

mse = mean_squared_error(y_test, y_pred1)
rmse = np.sqrt(mse)
print("RMSE Score :", rmse)

r2_score = r2_score(y_test, y_pred1)
print("R2 Score :",r2_score)
```

```python
# lets create a Random Forest Model

from sklearn.ensemble import RandomForestRegressor

model2 = RandomForestRegressor()
model2.fit(x_train, y_train)

y_pred2 = model2.predict(x_test)

# lets check the Model accuracy
from sklearn.metrics import r2_score, mean_squared_error

mse = mean_squared_error(y_test, y_pred2)
rmse = np.sqrt(mse)
print("RMSE Score :", rmse)

r2_score = r2_score(y_test, y_pred2)
print("R2 Score :",r2_score)
```
